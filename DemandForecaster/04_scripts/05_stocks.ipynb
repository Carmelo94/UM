```
import pandas_datareader.data as web
```

```
ticker_dict = {
    '^DJI':['Dow Jones Industrial Average', 'United States'],
    '^GSPC': ['S&P 500', 'United States'],
    '^IXIC': ['NASDAQ Composite', 'United States'],
    '^FTSE': ['FTSE 100', 'United Kingdom'],
    '^GSPTSE': ['S&P/TSX Composite index', 'Canada'],
    '^MXX': ['IPC Mexico', 'Mexico'],
    '^AORD': ['All Ordinaries', 'Australia']
}

tickers = list(ticker_dict.keys())

source = 'yahoo'
start_date = datetime(2017, 2, 1)
end_date = datetime(2021, 2, 28)

```

```
df = web.DataReader(tickers, source, start_date, end_date)
```

```
df_adj = df['Adj Close'].reset_index()
df_adj = df_adj.melt(id_vars='Date').dropna()
df_adj.sort_values(by=['Symbols', 'Date'], inplace=True)
df_adj['index_perc_change'] = df_adj.groupby('Symbols')['value'].transform(lambda x: x.pct_change())

df_adj['Ticker'] = df_adj['Symbols'].apply(lambda x: ticker_dict[x][0])
df_adj['market_name'] = df_adj['Symbols'].apply(lambda x: ticker_dict[x][1])

col_order = ['Date', 'Symbols', 'Ticker', 'market_name', 'value', 'index_perc_change']
df_adj = df_adj[col_order].reset_index(drop=True)
```
